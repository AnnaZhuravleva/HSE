```python
with open(r'belkin.txt', 'r', encoding = 'utf-8') as f:
    data = f.read()
with open(r'belkin0.txt', 'r', encoding = 'utf-8') as f:
    full = f.read()
data
```

```python
#попытки регулярных выражений
regex = '([А-Я]+[^.?!]*?[.?!]) ' #так появляются лишние пустые множества
my_split1 = re.split(regex, data)
my_split1
```

```python
regex = r'\W[^!.?]? ([А-Я][^.?!]*?[.?!]) ' #так исчезают точки...
my_split2 = re.split(regex, data)
my_split2
```

```python
regex = r'[.|!|?] [А-Я]' #так исчезают точки и начальные буквы...
my_split3 = re.split(regex, data)
my_split3
```

```python
my_split = re.split('@',(re.sub(r'([.?!])( )([А-Я])', r'\1@\3', data)))
my_split
```

```python
gold = [re.sub('[\n\t]', ' ', x) for x in sent_tokenize(data)]
gold
```

Мы видим довольно низкий показатель метрики:
    my_split не умееть делить по символу переноса строки '\n',

```python
trainer = PunktTrainer()
trainer.INCLUDE_ALL_COLLOCS = True
trainer.train('\n'.join(full))
 
tokenizer = PunktSentenceTokenizer(trainer.get_params())
```

```python
print(tokenizer._params.abbrev_types) #Видимо, у Пушкина сокращений нет)
```

```python
[re.sub('[\n\t]', ' ', x)[-10:] for x in tokenizer.tokenize(data)[:15]]
```

```python
gold_tokenized = tokenizer.tokenize(data)
print(len(gold_tokenized))
gold_tokenized
```

```python
tp = 0
fp = 0
fn = 0

for sent in gold_tokenized:
    if len(sent_tokenize(sent)) == 1:
        tp += 1
    else:
        fp += 1

for i in range(len(gold_tokenized)-1):
    sent1, sent2 = gold_tokenized[i], gold_tokenized[i+1]
    sent = ' '.join([sent1, sent2])
    if len(sent_tokenize(sent)) == 2:
        tp += 1
    else:
        fn += 1

precision = (tp/(tp+fp))
recall = (tp/(tp+fn))
f1 = 2*(precision*recall)/(precision+recall)
print('Precision - ', precision)
print('Recall - ', recall)
print('F1 - ', f1)
```

Итог:

  регулярное выражение ".!?" не позволяет добиться точного членения предложения, т.к. бывает прямая речь, бывают знаки табуляции, невидимые для данного регулярного выражения. Именно эти случаи и не увидел обученный на регулярном выражении my_split.
  
  Tokenizer разбил выделенный ему отрывок почти идеально, но не заметил случая с прямой речью (gold_tokenized).
